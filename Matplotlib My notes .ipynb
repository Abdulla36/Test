```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
#Funtional methood for plotting:

plt.plot(x,y,'r')    #same as matlab !!!!!!!!!!
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
#plt.show() for other IDE
```

```python
plt.subplot(1,2,1) #MATLAB BOOOYYY!!!!
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')  
```

```python
## Object Orinated Methood:

fig = plt.figure() #matlab!!!!!
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

fun fact: subplot automatically creates the fig.add_axes([]) for you and adjust based on the rows and columns input you add

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #[(left-right),(up-down),(width),(highet)]

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fix, axes = plt.subplots(3,3) #axes here adds 

plt.tight_layout() #helps space out the graphs, remove it and see the difference
```

```python
fix, axes = plt.subplots(1,2)
#the ,axes helps you deal with the plot as a list

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

### Figure Size, Aspect Ration and DPI 

```python
fig = plt.figure(figsize=(3,3)) # width and hight
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,3)) # width and hight

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```
