## distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill'])
# Safe to ignore warnings
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## matrix plot: heat map

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# we need to have varaibles on rows as well for this to work
# you dont want them to be python defult index for rows
```

#### cor method

```python
tc = tips.corr() # will take numeric only
tc
```

```python
sns.heatmap(tc) #colour scale :D
```

```python
sns.heatmap(tc,annot= True) 
```

```python
sns.heatmap(tc,annot= True,cmap='coolwarm') 
```

#### pivot table

```python
fp = flights.pivot_table(index ='month',columns ='year',values = 'passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=0.5) #:DDDDDDDDDDDDDDDDDDDDdd
```

## cluster map

```python
sns.clustermap(fp,cmap='coolwarm') #shows columns and rows that are similar to each others
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

Heat map will show data in the order you put them in 

Cluster map will show data so that similar groups are close to each other

## Grids 
### grids helps customise pairplot

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) #middle
g.map_upper(plt.scatter) #upper right
g.map_lower(sns.kdeplot)# lower right
```
